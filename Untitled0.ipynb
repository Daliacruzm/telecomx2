```python
df_clean = df.copy()
```

--- 1. Eliminar duplicados ---

```python
# Extract nested dictionary values into new columns
df_clean['gender'] = df_clean['customer'].apply(lambda x: x.get('gender'))
df_clean['SeniorCitizen'] = df_clean['customer'].apply(lambda x: x.get('SeniorCitizen'))
df_clean['Partner'] = df_clean['customer'].apply(lambda x: x.get('Partner'))
df_clean['Dependents'] = df_clean['customer'].apply(lambda x: x.get('Dependents'))
df_clean['tenure'] = df_clean['customer'].apply(lambda x: x.get('tenure'))

df_clean['PhoneService'] = df_clean['phone'].apply(lambda x: x.get('PhoneService'))
df_clean['MultipleLines'] = df_clean['phone'].apply(lambda x: x.get('MultipleLines'))

df_clean['InternetService'] = df_clean['internet'].apply(lambda x: x.get('InternetService'))
df_clean['OnlineSecurity'] = df_clean['internet'].apply(lambda x: x.get('OnlineSecurity'))
df_clean['OnlineBackup'] = df_clean['internet'].apply(lambda x: x.get('OnlineBackup'))
df_clean['DeviceProtection'] = df_clean['internet'].apply(lambda x: x.get('DeviceProtection'))
df_clean['TechSupport'] = df_clean['internet'].apply(lambda x: x.get('TechSupport'))
df_clean['StreamingTV'] = df_clean['internet'].apply(lambda x: x.get('StreamingTV'))
df_clean['StreamingMovies'] = df_clean['internet'].apply(lambda x: x.get('StreamingMovies'))

df_clean['Contract'] = df_clean['account'].apply(lambda x: x.get('Contract'))
df_clean['PaperlessBilling'] = df_clean['account'].apply(lambda x: x.get('PaperlessBilling'))
df_clean['PaymentMethod'] = df_clean['account'].apply(lambda x: x.get('PaymentMethod'))
df_clean['MonthlyCharges'] = df_clean['account'].apply(lambda x: x.get('Charges', {}).get('Monthly'))
df_clean['TotalCharges'] = df_clean['account'].apply(lambda x: x.get('Charges', {}).get('Total'))

# Now drop duplicates based on the new columns (excluding the original dictionary columns)
columns_to_check = [col for col in df_clean.columns if col not in ['customer', 'phone', 'internet', 'account']]
df_clean = df_clean.drop_duplicates(subset=columns_to_check)
```

--- 2. Normalizar valores nulos ---
Reemplazar espacios en blanco o cadenas vacías por NaN

```python
df_clean = df_clean.replace(r'^\s*$', pd.NA, regex=True)
```



--- 3. Convertir columnas numéricas ---
Forzamos la conversión de "account.Charges.Total" a numérico

```python
df_clean["TotalCharges"] = pd.to_numeric(
    df_clean["TotalCharges"], errors="coerce"
)
```

#### --- 4. Normalizar categorías ---
#### Eliminamos registros donde Churn está vacío

```python
df_clean = df_clean[df_clean["Churn"].notna()]
```

#### Reemplazamos "No internet service" por "No"

```python
internet_cols = [col for col in df_clean.columns if col.startswith("internet.")]
for col in internet_cols:
    df_clean[col] = df_clean[col].replace("No internet service", "No")
```

#### Reemplazamos "No phone service" por "No"

```python
df_clean.loc[:, "MultipleLines"] = df_clean["MultipleLines"].replace(
    "No phone service", "No"
)
```

#### --- 5. Verificar limpieza ---

```python
print("Valores nulos después:", df_clean.isnull().sum().sum())
columns_to_check = [col for col in df_clean.columns if col not in ['customer', 'phone', 'internet', 'account']]
print("Duplicados después:", df_clean.duplicated(subset=columns_to_check).sum())
print("Valores únicos en Churn:", df_clean["Churn"].unique())
print("Tipos de datos:\n", df_clean.dtypes)
```
